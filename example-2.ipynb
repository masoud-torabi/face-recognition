```python
from ultralytics import YOLO
import cv2
from deep_sort_realtime.deepsort_tracker import DeepSort
import torch
```

```python
torch.cuda.is_available()
```

```python
detector = YOLO("yolov8n-face.pt")
tracker = DeepSort(
    max_age=30,
    n_init=3,
    max_iou_distance=0.7
)

cap = cv2.VideoCapture(0)

assert cap.isOpened(), "Cannot open video"
while True:
    ret, frame = cap.read()
    if not ret:
        break

    detections = []

    results = detector(frame, conf=0.3)[0]

    for box in results.boxes:
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
        conf = float(box.conf[0])
        detections.append(([x1, y1, x2 - x1, y2 - y1], conf, "face"))

    tracks = tracker.update_tracks(detections, frame=frame)

    for track in tracks:
        if not track.is_confirmed():
            continue

        track_id = track.track_id
        x, y, w, h = map(int, track.to_ltrb())

        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame, f"Person {track_id}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    cv2.imshow("Face Tracking", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
